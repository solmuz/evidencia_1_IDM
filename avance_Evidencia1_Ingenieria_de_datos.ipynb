#### Ejercicio 16: Manejo de valores faltantes
**Objetivo:** Detectar y tratar los valores faltantes en el DataFrame.

```python
# Verificar valores nulos
dataFrame.isnull().sum()
```

```python
# Rellenar valores faltantes con la media en columnas numericas
numeric_columns = dataFrame.select_dtypes(include=[np.number]).columns
dataFrame[numeric_columns] = dataFrame[numeric_columns].fillna(dataFrame[numeric_columns].mean())
```

#### Ejercicio 18: Uso de `merge` para combinar DataFrames
**Objetivo:** Demostrar cómo combinar dos DataFrames usando una clave común.

```python
# Crear un DataFrame adicional con información extra
dataFrame2 = pd.DataFrame({
    'age': [30, 40, 50],
    'extra_info': ['A', 'B', 'C']
})
merged_data = dataFrame.merge(dataFrame2, on='age', how='left')
print(merged_data)
```

#### Ejercicio 22: Crear un reporte de correlación entre variables
**Objetivo:** Calcular y visualizar la matriz de correlación para entender las relaciones entre variables.

```python
# Filtrar solo las columnas numéricas antes de calcular la correlación
correlacion = dataFrame.select_dtypes(include=['number']).corr()

# Crear el heatmap con la matriz de correlación corregida
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
